## Section 3 -- Tool Demo: Real Agents in Action (15 min)

### Demo 1: ChatGPT Agent Mode

**What it is:** When you give ChatGPT a complex task, it can autonomously decide to search the web, run Python code, generate images, read uploaded files, and chain these actions together — without you prompting each step.

**Live demo task:**
```
Find the top 5 most-starred Python repositories on GitHub created
in the last 30 days. For each, give me the repo name, description,
star count, and primary language breakdown. Present as a table.
```

**What to watch for (map it to the loop):**

```
THINK  → "I need to search GitHub for recent popular repos"
ACT    → [Searches the web]
OBSERVE→ Gets search results, but they're not in table format
THINK  → "I should run Python to parse and format this data"
ACT    → [Runs Python code to create a formatted table]
OBSERVE→ Table is ready
DONE   → Returns the formatted table to the user
```

> "Notice: the user didn't say 'search the web' or 'run Python code.' The agent **decided** to use those tools based on the task. That's autonomy."

**Beginner Note:** You've probably already seen ChatGPT do this — now you know it's following the agent loop.

### Demo 2: Claude Computer Use

**What it is:** Claude can see your screen, move the mouse, click buttons, type text, and interact with any application on your computer — like a remote-controlled colleague sitting at your desk.

**Live demo task:**
```
Open my browser, go to GitHub, find the FastAPI repository,
navigate to the Issues tab, and summarise the 3 most recent
open issues.
```

**What to watch for (map it to the loop):**

```
THINK  → "I need to open the browser first"
ACT    → [Takes screenshot, clicks browser icon]
OBSERVE→ Browser is open, I can see the address bar
THINK  → "Now I need to navigate to GitHub"
ACT    → [Types github.com/tiangolo/fastapi in address bar, presses Enter]
OBSERVE→ FastAPI repo page loaded
THINK  → "I need to click the Issues tab"
ACT    → [Clicks Issues tab]
OBSERVE→ Issues page loaded, I can see issue titles
THINK  → "I need to read the top 3 open issues"
ACT    → [Reads issue titles and descriptions from the page]
OBSERVE→ I have the information
DONE   → Returns summary of the 3 issues
```

> "This is the most visceral demonstration of an agent. You can literally watch it move the mouse. Every click is an ACT, every screenshot is an OBSERVE, and between each pair the LLM is THINKing about what to do next."

**Intermediate Note:** Claude computer use is currently in beta. The latency is noticeable — each loop iteration involves a screenshot, LLM reasoning, and an action, which takes several seconds.

**Advanced Extension:** Under the hood, Claude computer use works by taking screenshots (OBSERVE), processing them as images (THINK), and outputting coordinates + actions (ACT). It's the exact same ReAct loop, just with a computer screen as the environment.

### Demo 3: n8n AI Agent Node

**What it is:** n8n is a visual workflow automation platform (like Zapier, but more powerful and self-hostable). Its AI Agent node lets you build agent loops visually — you can see the Think-Act-Observe cycle as actual nodes on a canvas.

**Live demo task:**
```
Build an n8n workflow where an AI agent:
1. Receives a topic via webhook
2. Searches the web for recent articles
3. Reads the top 3 articles
4. Writes a summary
5. Sends the summary via email
```

**What to watch for:**

```
n8n Canvas:

  [Webhook Trigger] → [AI Agent Node] → [Send Email]
                            │
                     Connected tools:
                     ├── Web Search tool
                     ├── Web Page Reader tool
                     └── Text Summariser tool

The AI Agent node IS the ReAct loop:
  - The LLM inside it THINKs about which tool to call
  - It ACTs by calling one of the connected tools
  - It OBSERVEs the tool's output
  - It decides if it needs another tool call or if it's done
```

> "n8n makes the agent architecture visible. You can literally see the tools as separate nodes connected to the agent. The agent decides at runtime which tools to use and in what order. You didn't hardcode the sequence — the AI figures it out."

**Beginner Note:** n8n is visual — no code required. Think of it as drawing a flowchart that actually runs.

### Exercise 4.2 -- The Failure Case: When the Loop Goes Wrong

**Give ChatGPT a task that's likely to trigger agent failure modes:**

```
Find the exact number of commits made to the Linux kernel
repository on GitHub in the last 24 hours. Give me the precise count.
```

**Watch for these common failure patterns:**

```
FAILURE MODE 1 — Infinite loop:
  Agent keeps searching, never finds a satisfying answer,
  keeps trying different queries.
  WHY: The stop condition is too strict ("exact number")
  and live data keeps changing.

FAILURE MODE 2 — Hallucinated tool use:
  Agent claims to have called the GitHub API but actually
  just made up a number.
  WHY: The agent doesn't have a real GitHub API tool, so
  it substitutes a guess.

FAILURE MODE 3 — Premature stop:
  Agent gives an approximate answer when you asked for exact.
  WHY: After a few failed attempts, it decides "close enough"
  and stops the loop.
```

> "Agent failure modes are just as important to understand as agent successes. In production, you'll need to handle all three: loop limits to prevent infinite runs, tool validation to catch hallucinated actions, and clear success criteria so the agent knows when 'good enough' is actually good enough."

## Section 5 -- Guided Exercise: Anatomy of a Real Agent (20 min)

> "Now let's reverse-engineer a real agent. I'll describe an agent that exists in production, and we'll break down its architecture together — mapping every component to our three pillars."

### Case Study: An Automated Bug Triage Agent

**Scenario:** A mid-size engineering team gets 50+ bug reports per day in their issue tracker. They've built an agent that automatically triages new bugs — categorises them, assigns priority, routes them to the right team, and adds initial diagnostic notes.

**Let's map the architecture:**

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/191/589/original/5.png?1776675423)



### Decompose the Agent

**Fill in this framework for the bug triage agent:**

````python
# AGENT DECOMPOSITION FRAMEWORK
# Fill this in based on the case study above.

agent_spec = {
    "name": "Bug Triage Agent",

    "goal": "Automatically categorise, prioritise, and route new bug reports",

    "trigger": "New issue created in issue tracker",

    "tools": [
        "Codebase search — find files related to error traces",
        "Deployment history — check recent deploys for regressions",
        "Issue tracker API — update priority, assignee, comments",
        # What other tools might be useful?
        # "Slack API — notify the assigned team?",
        # "Log search — find related error patterns?",
    ],

    "memory": [
        "List of teams and their code ownership areas",
        "Recent deployment history",
        "Past triage decisions (for learning patterns)",
    ],

    "success_criteria": [
        "Issue has a priority assigned (P0-P3)",
        "Issue is routed to a specific team",
        "A diagnostic comment is added with initial analysis",
    ],

    "failure_modes": [
        "Bug report is too vague — no error trace, no steps to reproduce",
        "Error is in a new service with no ownership mapping",
        "Agent assigns wrong priority (P3 when it's actually P0)",
    ],

    "guardrails": [
        "Never auto-close an issue — only triage",
        "If confidence is low, flag for human review instead of guessing",
        "Maximum 3 tool calls per issue to prevent runaway loops",
    ],
}

# Print the spec
for key, value in agent_spec.items():
    print(f"\n{key.upper()}:")
    if isinstance(value, list):
        for item in value:
            print(f"  - {item}")
    else:
        print(f"  {value}")
````

> "Notice the 'guardrails' section. This is critical. An agent without guardrails is like a script with no error handling — it WILL do something unexpected eventually. The guardrails are how you keep it safe."

**Intermediate Note:** In production, the 'confidence is low' guardrail is usually implemented by asking the LLM to self-rate its confidence on a scale, then routing to a human if it's below a threshold.

**Advanced Extension:** This pattern is called 'human-in-the-loop.' It's the most common production agent pattern — the agent handles the easy 80% autonomously, and escalates the hard 20% to a human.

---

### Key takeaways

1. **An agent is a loop, not a single call.** The Think-Act-Observe cycle is the fundamental pattern. Every agent you'll ever build or use follows this structure.

2. **Tools are what make agents useful.** Without tools, an agent is just a chatbot that talks to itself. Tools let it search, compute, read files, call APIs, and interact with the real world.

3. **Memory is what makes agents coherent.** Without memory, each loop iteration would forget what happened before. Memory tracks progress, context, and intermediate results.

4. **Autonomy is what makes agents agents.** The LLM decides what to do next — not the user, not a hardcoded script. This is both the power and the risk.

5. **Guardrails are not optional.** Agents without boundaries will loop forever, take wrong actions, or confidently do the wrong thing. Design guardrails into every agent spec.

6. **Spec before you code.** The biggest mistake in agent development is jumping to code before designing the loop, tools, success criteria, and failure modes. Your hand-drawn spec IS the architecture.
