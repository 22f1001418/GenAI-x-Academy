### 1. What is AI? (5 min)

**Instructor says:**

> "AI stands for Artificial Intelligence. It's a broad term for software that does things we'd normally call 'intelligent' -- recognising images, understanding language, making decisions, playing chess."

**The key distinction:**


### 3. What is an LLM? (6 min)

> "LLM stands for **Large Language Model**. It's the engine behind ChatGPT, Claude, Gemini, and most of the tools we'll use in this course. Here's what you need to know -- and nothing more."

**An LLM is an AI trained on massive amounts of text to predict the next word.**

That's the entire core idea. You give it text, it continues the text.

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/194/281/original/2.png?1777875226)


**Why "Large"?**
- Billions of internal parameters (think of them as adjustable dials)
- Trained on internet-scale text (books, websites, code, articles)
- "Large" is what makes them capable of handling such a wide range of tasks

**The major LLMs you'll see in this course:**

```
┌──────────────┬──────────────────┐
│ LLM          │ Made by          │
├──────────────┼──────────────────┤
│ GPT-4o, o3   │ OpenAI           │
│ Claude       │ Anthropic        │
│ Gemini       │ Google           │
│ Llama        │ Meta (open)      │
└──────────────┴──────────────────┘
```

**One thing to know for later:** Modern LLMs aren't text-only anymore. They can also process images, audio, and video (this is called "multimodal"). But the core mechanism -- predict the next token -- is still text prediction.

> "Every single class in this course uses LLMs. You don't need to know how they work internally -- just what they do: you give them text, they give back text."

### 4. What is an AI Agent? (6 min)

> "This is the term you'll hear most in the second half of the course. Let me give you just enough to recognise it -- we'll go deep in later classes."

**An AI Agent is a system that doesn't just answer questions -- it completes tasks.**

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/194/282/original/3.png?1777875321)


**Concrete example:**

```
"Book a meeting with Sarah next week."

  Chatbot says: "You can send Sarah a calendar invite for
                 next Tuesday at 2pm."
                 (gives advice, YOU do the work)

  Agent does:    1. Reads your calendar for free slots
                 2. Reads Sarah's calendar for free slots
                 3. Finds a mutual time
                 4. Drafts a calendar invite
                 5. Sends it
                 6. Tells you: "Done. Meeting booked for
                    Tuesday 2pm."
                 (does the work FOR you)
```

**The three things that make an agent an agent:**

```
              ┌──────────┐
              │   LLM    │  (the brain -- reasons and decides)
              │  (Brain) │
              └────┬─────┘
                   │
         ┌────────┼────────┐
         │        │        │
         ▼        ▼        ▼
    ┌────────┐ ┌────────┐ ┌──────────┐
    │ TOOLS  │ │ MEMORY │ │ AUTONOMY │
    │        │ │        │ │          │
    │ Search │ │Remembers│ │ Decides  │
    │ Email  │ │ what   │ │ next step│
    │ Code   │ │ it did │ │ on its   │
    │ APIs   │ │        │ │ own      │
    └────────┘ └────────┘ └──────────┘
```

> "We'll spend half this course building agents. For now, just remember: a chatbot answers, an agent **does**."

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/194/283/original/4.jpeg?1777875713)

### 8. JSON Basics (7 min)

> "When your code talks to an AI API, the response comes back in a format called JSON. It's the universal language of API responses. Let me show you how to read it -- you don't need to write it from scratch."

**JSON = key-value pairs inside curly braces.**

````python
# JSON looks like this:
# (This is a Python dictionary -- same syntax as JSON)

# Simple JSON:
user = {
    "name": "Priya",
    "age": 24,
    "is_student": True
}

print("Name:", user["name"])
print("Age:", user["age"])
````

````python
# JSON can be NESTED (objects inside objects):

api_response = {
    "status": "success",
    "data": {
        "message": "Hello! How can I help you today?",
        "model": "gemini-2.5-flash",
        "tokens_used": 42
    },
    "choices": [
        {"text": "Option A"},
        {"text": "Option B"}
    ]
}

# Reading nested values:
print("Message:", api_response["data"]["message"])
print("Model:",   api_response["data"]["model"])
print("First choice:", api_response["choices"][0]["text"])
````

**The JSON rules that matter:**

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/194/287/original/8.jpeg?1777877215)


**What success looks like:** You can look at a JSON blob and find the value you need. That's all you need for this course.
