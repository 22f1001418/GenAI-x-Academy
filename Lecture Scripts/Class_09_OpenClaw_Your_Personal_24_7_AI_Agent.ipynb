## Section 2 -- Concept in 5 Minutes: How OpenClaw Works

> "Let me show you the architecture. Once you see this diagram, everything about OpenClaw clicks."

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/192/944/original/1.jpeg?1777286568)


**The key pieces:**
- **Gateway** -- the core process. Runs on your machine, handles all communication
- **Skills** -- pluggable capabilities (web search, file ops, shell commands, email). Think of them as the tools from Class 5's agent framework
- **Memory** -- stored as plain Markdown files on your disk. No cloud, no database. You can read them in any text editor
- **Heartbeat** -- a scheduler that wakes the agent up periodically to check on long-running tasks. This is what makes it "24/7" -- it can work while you're away
- **LLM API** -- OpenClaw doesn't include a model. You connect YOUR choice of LLM: Gemini (free), Claude, OpenAI, or a local model via Ollama
- **Messaging** -- you talk to OpenClaw through apps you already have. It's not another UI to learn

````python
# TERMINAL COMMANDS — Run these in order.
# This is the Docker install method (recommended for security).

# Step 1: Clone the OpenClaw repository
# git clone https://github.com/openclaw/openclaw.git

# Step 2: Navigate into the directory
# cd openclaw

# Step 3: Run the Docker setup script
# export OPENCLAW_IMAGE="ghcr.io/openclaw/openclaw:latest"
# ./scripts/docker/setup.sh

# That's it. Three commands.
# The setup script will:
#   1. Build the OpenClaw Docker image
#   2. Run the onboarding wizard (in terminal)
#   3. Generate a gateway token
#   4. Start the gateway via Docker Compose

# When it finishes, open your browser to:
#   http://127.0.0.1:18789
# This is the OpenClaw Control UI dashboard.
````

### Exercise 5.1 -- Build the n8n Webhook Workflow

**Open your n8n instance from Class 6 (or create a new one at [app.n8n.cloud](https://app.n8n.cloud/)).**

**Build this workflow:**

**Step 1 -- Webhook Trigger:**
- Add a **Webhook** node (not Chat Trigger -- this time we want a URL endpoint)
- Set method to **POST**
- Note the webhook URL (it looks like `https://your-n8n.app.n8n.cloud/webhook/xxxxx`)

**Step 2 -- Parse the incoming data:**
- The webhook will receive JSON from OpenClaw with fields like `topic`, `summary`, and `recipient`

**Step 3 -- Google Docs action:**
- Add a **Google Docs > Create Document** node
- Set the title to: `Research: {{ $json.topic }}`
- Set the content to: `{{ $json.summary }}`

**Step 4 -- Gmail action (optional):**
- Add a **Gmail > Send Email** node
- Set recipient to: `{{ $json.recipient }}`
- Set subject to: `Research Summary: {{ $json.topic }}`
- Set body to: `{{ $json.summary }}`

**Step 5 -- Respond to Webhook:**
- Add a **Respond to Webhook** node
- Return: `{"status": "success", "message": "Document created and email sent"}`

**Activate the workflow** (toggle it ON in the top bar).

````python
# YOUR n8n WORKFLOW SHOULD LOOK LIKE THIS:
#
#  [Webhook] --> [Google Docs: Create] --> [Gmail: Send] --> [Respond to Webhook]
#     POST          Title: topic            To: recipient     {"status": "success"}
#     receives      Content: summary        Body: summary
#     JSON
#
# The webhook URL is what OpenClaw will call.
# Credentials for Google Docs and Gmail live in n8n -- not in OpenClaw.

# Test the webhook manually first (before connecting OpenClaw):
# In a terminal or browser tool like Postman, send:

# curl -X POST https://your-n8n.app.n8n.cloud/webhook/xxxxx \
#   -H "Content-Type: application/json" \
#   -d '{
#     "topic": "Test Topic",
#     "summary": "This is a test summary from the terminal.",
#     "recipient": "your-email@example.com"
#   }'

print("If the curl command creates a Google Doc and sends an email,")
print("your n8n workflow is ready. Now we connect OpenClaw to it.")
````

**What should happen:**

2. OpenClaw writes a summary (LLM reasoning)
3. OpenClaw calls the n8n webhook (simple HTTP call -- no LLM needed)
4. n8n creates the Google Doc (using its own credentials)
5. n8n sends the email (using its own Gmail credentials)
6. n8n responds to OpenClaw with success

**What success looks like:** A Google Doc appears in your Drive, an email arrives in your inbox, and the agent confirms via Telegram -- all from one message.

**Intermediate Note:** Notice the security improvement. Your Gmail credentials are in n8n, not in OpenClaw's environment. A prompt injection against OpenClaw can't steal them.

**Advanced Extension:** The n8n webhook pattern is infinitely extensible. You can build n8n workflows that talk to Slack, Notion, databases, CRMs -- and OpenClaw calls them all through simple webhooks, never seeing any credentials.
