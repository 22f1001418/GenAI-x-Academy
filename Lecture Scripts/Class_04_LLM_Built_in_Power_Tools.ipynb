## Section 3 — Tool Demo: Building a Gemini Gem (15 min)

> "Let's start with Gemini Gems — because they're free, fast, and the simplest to set up. Watch me build one, then you'll build your own."

### What is a Gem?

A Gem is a custom version of Gemini that you configure once and reuse forever. Instead of re-pasting your guidelines, re-stating your persona, and re-uploading your reference files every session, you bake all of that into a dedicated assistant that remembers its job every time you open it.

**And it's completely free.** Anyone with a Google account can create and use Gems.

### When to build a Gem vs. just prompting:

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/191/324/original/4.png?1776496165)



### Step-by-step demo (share screen):

**Step 1 — Open Gem Manager.**  
Go to [gemini.google.com](https://gemini.google.com/app). In the left sidebar, click **"Gem manager"**. You'll see some pre-made Gems from Google plus a **"+ New Gem"** button. Click it.

**Step 2 — Name and describe your Gem.**  
The creation screen has fields for name, description, and a large instructions box, with a preview chat panel on the right.

For the demo, we'll build a **"Code Review Bot"** — a SWE-focused example:
- **Name:** `Code Review Bot`
- **Description:** `A strict code reviewer that checks Python code for bugs, security issues, and style violations.`

**Step 3 — Write the instructions.**  
This is the heart of the Gem — a permanent system prompt. Everything from Class 3 about system prompts applies here, except now it **persists across every conversation** with this Gem.

````python
# Gem Instructions — "Code Review Bot"

# You are a senior Python code reviewer at a top-tier tech company.
#
# ## Your Review Process
# For every code snippet shared, you must:
# 1. Check for bugs and logical errors
# 2. Flag security vulnerabilities (SQL injection, hardcoded secrets,
#    unsafe deserialization, etc.)
# 3. Note PEP 8 style violations
# 4. Rate readability on a scale of 1-5
# 5. Suggest the single most impactful refactor
#
# ## Output Format
# Use this structure for every review:
# - **Bugs:** (list or "None found")
# - **Security:** (list or "No issues")
# - **Style:** (top 3 violations or "Clean")
# - **Readability:** X/5 with one-line justification
# - **Top Refactor:** (one specific suggestion with before/after code)
#
# ## Constraints
# - Keep reviews under 300 words
# - Be direct and constructive — no fluff
# - If the code is good, say so briefly and move on
# - Always suggest type hints if they're missing
````

>"Notice what we've done here. We took our Class 3 system prompt knowledge and made it *permanent*. This Gem will follow these rules in every conversation — no re-pasting needed."

**Pro tip:** If you're not sure how to write detailed instructions, type a simple one-sentence goal and click the **magic pencil icon** at the bottom of the instructions box. Gemini will auto-expand it into structured instructions you can then edit.

**Step 4 — Upload knowledge files.**  
Below the Instructions field, there's a **"Knowledge"** section. Click **"Upload files"** and add any reference documents you want the Gem to access. You can also connect files directly from Google Drive.

For our Code Review Bot, we might upload:
- A team style guide document
- A list of banned patterns or known anti-patterns
- An architecture decision record (ADR) for context

Once uploaded, these files become **permanent context**. Every conversation with this Gem can draw on them.

> A quick note: the mechanism by which the Gem searches through those uploaded documents and pulls relevant sections is a technique called **RAG (Retrieval Augmented Generation)**. We'll dedicate Class 5 to understanding that. For today, what matters is the practical outcome: you upload files, and the Gem references them.

**Step 5 — Preview and save.**  
On the right side, there's a Preview pane. Test your Gem before saving. Once satisfied, click **Save**. The Gem appears in your sidebar, ready to reuse.


**Step 6 — Testing it with a code, having some flaws**  

````python
## paste the following code as an input to the code review gem you just created

import random
import math as m
import datetime

def calcAverage(numbers):
    total = 0
    for i in numbers:
        total += i
    return total / len(numbers)

def generate_numbers(n):
    nums = []
    for i in range(0, n):
        nums.append(random.randint(1, 100))
    return nums

def printReport(data):
    avg = calcAverage(data)
    print("Average is:", avg)
    print("Max value is:", max(data))
    print("Min value is:", min(data))

def weird_function(x,y):
    if x > y:
        return x - y
    else:
        return y - x

def main():
    count = 10
    numbers = generate_numbers(count)

    printReport(numbers)

    # inconsistent naming + unused variable
    currentTime = datetime.datetime.now()
    unused_var = "hello"

    for i in range(len(numbers)):
        if numbers[i] % 2 == 0:
            print("Even:", numbers[i])
        else:
            print("Odd:", numbers[i])

    # inconsistent spacing and logic
    result=weird_function(5 ,10)
    print("Result:",result)

    # unnecessary complexity
    squares = list(map(lambda x: x**2, numbers))
    print("Squares:", squares)

    # possible division by zero (edge case ignored)
    empty_list = []
    try:
        print(calcAverage(empty_list))
    except:
        print("Error calculating average")

if __name__ == "__main__":
    main()
````

### Exercise 5.1 — Temperature Experiment

**Open [Google AI Studio](https://aistudio.google.com/). Click "Create new prompt".**

**System Instructions (paste this):**
```
You are a backend code generator. Write clean, production-ready Python.
Always include error handling and type hints.
```

**User prompt (use the same prompt for all three runs):**
```
Write a function that fetches user data from a REST API,
retries on failure, and returns a parsed User object.
```

**Run this prompt THREE times, changing only the temperature:**

| Run | Temperature | What to observe |
|:----|:-----------|:----------------|
| Run 1 | **0.1** | Straightforward, conventional approach. Likely uses `requests` + simple retry loop |
| Run 2 | **0.7** | More varied. Might use `tenacity`, `httpx`, or add more sophisticated error handling |
| Run 3 | **1.2** | Creative but potentially unpredictable. May invent unusual patterns or use uncommon libraries |

**Record your observations:**
- Which run produced the most reliable, production-ready code?
- Which run was most creative or surprising?
- At what point did the output start to feel unreliable?



**What success looks like:** You should notice a clear difference between runs. Low temperature = safe and predictable. High temperature = creative but potentially less reliable. There's no single "right" setting — it depends on your use case.

### What you're likely to see at each temperature:

 Temperature 0.1 — Conservative
 - Uses requests library (most common choice)
 - Simple for loop for retries
 - Basic try/except
 - Predictable, safe, boring

 Temperature 0.7 — Balanced
 - Might use httpx or aiohttp
 - May add exponential backoff
 - Better error classification
 - Good balance of quality and creativity

 Temperature 1.2 — Creative
 - Might use tenacity, backoff, or custom decorators
 - Could add circuit breaker pattern
 - May suggest unusual but interesting approaches
 - Verify carefully — creativity != correctness
