## Section 2 — The Three Steps of RAG

> "RAG sounds fancy. The actual idea is so simple you could explain it to a 10-year-old: **give the AI the right notes before it answers.** That's the whole pattern. Everything else is plumbing."

### The three steps — in one picture:

### One-line analogies for the three tools we'll use today:

- **NotebookLM** = a specialist who only reads from *your bookshelf*
- **Perplexity** = a researcher who searches the *live web* and cites every source
- **NoteGPT** = a note-taker who reads a *single long video* and answers from its transcript

**Beginner Note:** All three are **no-code** — you upload or paste, then type questions. No programming required.

**Intermediate Note:** In a later class, you'll call an LLM API and pass context yourself — that's "RAG by hand." It looks exactly like Step 2 above, just in Python.

### The shape of a RAG call — in 8 lines

Before any tools, here's what RAG *looks like* in code. No API key, no network call — just the mental model. This is what every RAG tool is doing behind the scenes.

```python
# The universal shape of every RAG call

def ask_llm(question, context=None):
    prompt = f"Context:\n{context}\n\nQuestion: {question}" if context else question
    print("──── PROMPT SENT TO THE LLM ────")
    print(prompt)

# Without RAG — the LLM has nothing to go on except its training data
ask_llm("What's our company's refund policy?")

print()

# With RAG — real policy text stuffed in front of the question
ask_llm(
    "What's our company's refund policy?",
    context="Acme Corp refund policy (2026): full refund within 30 days of purchase, store credit after 30 days up to 90 days, no refunds after 90 days."
)
```

**What you'll see:** two printed prompts. The second one has the policy text **stuffed in front of the question**. That's literally all RAG is. When NotebookLM answers a question about your PDF, it's doing that exact operation — just with your uploaded content as the context string.

## Section 3 — Tool Demo: NotebookLM Deep Dive (20 min)

> "Let's start with NotebookLM — because it's free, requires no setup, and is genuinely the best way to feel what 'grounded AI' means. Watch me build a notebook, then you'll build your own."

### What is NotebookLM?

NotebookLM is Google's grounding tool. You upload sources — PDFs, Google Docs, websites, YouTube URLs, even raw pasted text — and it builds a private AI assistant that answers **only** from those sources, with citations back to the exact passage.

Unlike ChatGPT, it has **no access** to the open web or its general training data when answering your questions. If the answer isn't in your sources, it says so. That "I don't know" is the feature, not a bug.

**And it's completely free** with a Google account.

### When to use NotebookLM vs just prompting:

| Use NotebookLM when... | Just prompt ChatGPT/Claude when... |
|:---|:---|
| You have specific documents you want answers from | You want general knowledge or creative help |
| You need citations you can verify | You're brainstorming or drafting |
| The topic is domain-specific (your course, your company) | The topic is broad and well-known |
| Answers must not be made up | Approximate is okay |

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/191/360/original/5.png?1776507470)


### Step-by-step demo (share screen):

**Step 1 — Open NotebookLM.**
Go to [notebooklm.google.com](https://notebooklm.google.com) and sign in with any Google account. You'll see a dashboard of your past notebooks (empty for first-time users) and a **"+ Create new"** button.

**Step 2 — Create a new notebook.**
Click **"+ Create new"**. You'll immediately be prompted to add sources. **This is the most important step** — without sources, the notebook has nothing to ground in.

**Step 3 — Add sources (the demo).**
Click **"+ Add sources"**. You'll see options for: PDF/file upload, Google Drive, Website URL, YouTube URL, pasted text.

For the demo, we'll build a **"System Design Study Notebook"**. Add at least three different source types:
- 📄 **PDF:** a chapter from a system design textbook (or any PDF you have)
- 🔗 **Website:** a blog post like `https://highscalability.com/...` or official docs
- 🎥 **YouTube:** paste a URL for a talk like *"System Design Interview"* — NotebookLM will transcribe and index it

Wait ~30 seconds per source for processing. You'll see a little green checkmark when each is ready.

**Step 4 — Ask the first question.**
In the chat box, ask something you **already know the answer to** from your sources. This lets you verify the AI isn't bluffing.

Example question for our System Design notebook:
- *"What's the difference between horizontal and vertical scaling according to my sources?"*

**Pro tip:** Start questions with *"According to my sources..."* or *"Based on what I uploaded..."* — it keeps the model anchored and makes citation-checking easier.

**Step 5 — Check the citations.**

Every answer comes with numbered citations like `[1]`, `[2]`, `[3]`. Click one. NotebookLM will jump to the **exact sentence** in the exact source it came from.

This is the feature. Without it, you're back to hallucinations.

**Step 6 — Test the refusal behaviour.**

Now ask something **not in your sources**:
> *"Who won the IPL final in 2026?"*

Watch NotebookLM politely decline: *"The sources don't contain information about that topic."*

> "A grounded AI that refuses is more valuable than a clever one that lies. That refusal is the whole point."

**Step 7 — Generate the power features.**

On the right panel, NotebookLM offers four auto-generated artifacts from your sources:
- 📘 **Study guide** — key terms, questions, essay prompts
- 📋 **Briefing doc** — cross-source synthesis
- ❓ **FAQ** — common questions with grounded answers
- 🎧 **Audio Overview** — two AI hosts discuss your sources in a podcast format (genuinely uncanny — click it, wait 60 seconds, listen)

**Beginner Note:** Don't overthink the sources — a 10-page PDF and two URLs is plenty to start.

**Intermediate Note:** Try uploading sources that partially overlap in content but disagree on specifics. Then ask a disagreement question. The answer will reveal how well NotebookLM synthesises conflicting sources.

**Advanced Note:** NotebookLM caps at 50 sources per notebook. Upload 15–20. Then ask a synthesis question spanning 5+ sources. Notice retrieval quality degrading — that's a real trade-off you'll hit in production RAG systems too.

```python
# Starter Question Templates for NotebookLM
# Copy these patterns, replace the bracketed parts, paste into NotebookLM.

# 1. FACTUAL (grounded lookup)
# According to my sources, what is [concept]?

# 2. COMPARISON (cross-source)
# Where do [source 1] and [source 2] disagree about [topic]?

# 3. SYNTHESIS (multi-source chain reasoning)
# Based on all my sources, what are the three most important
# things a beginner should know about [topic]? Cite which
# source each point comes from.

# 4. EXPLAIN-LIKE-I'M-FIVE (grounding + simplification)
# Using only my uploaded sources, explain [complex concept]
# as if I've never heard of it before.

# 5. DEVIL'S ADVOCATE (probe weaknesses)
# What are the strongest objections to the main argument
# in my sources? Cite which source raises each concern
# (or note if none do).
```

## Section 5 — Guided Exercise: Grounded vs Ungrounded (20 min)

> "Time to feel the hallucination difference in your bones. You'll ask the *same question* to three tools and observe the response. The goal isn't to decide which tool is 'best' — it's to understand what grounding actually changes."

### Setup — three tabs open:
1. 🤖 **ChatGPT** at [chat.openai.com](https://chat.openai.com) — **with web search turned OFF** (click the tools icon, disable web search)
2. 🌐 **Perplexity** at [perplexity.ai](https://www.perplexity.ai)
3. 📚 **Your NotebookLM** notebook from Section 3

### Exercise 5.1 — The Same Question, Three Tools

Pick **one** question. Ask it verbatim in all three tabs.

```python
# Test Prompts for the Grounded vs Ungrounded Exercise
# Pick ONE. Ask all three tools the same question.

# Prompt A — Time-sensitive (best shows the freshness gap)
# What are the three biggest changes in Next.js in the last 6 months?

# Prompt B — Niche/specific (best shows the hallucination gap)
# What are the most common production failure modes of
# Kubernetes pods and their typical root causes?

# Prompt C — Your-source-specific (only NotebookLM can answer)
# Based on my uploaded sources, what are the three key takeaways
# from the material I've studied?
```

**What to observe:** Don't just fill it — *notice the feel.* Which one sounded most confident? Was confidence correlated with correctness? Which one was willing to say "I don't know"?

### Exercise 5.2 — The False Premise Test

This one reveals which tools will lie to please you. Ask all three:
> *"Explain why Rust uses garbage collection for memory safety."*

(Rust famously does **not** use garbage collection — it uses ownership and borrowing.)

**What to observe:**
- Which tool **pushes back** and corrects the premise?
- Which just **plays along** and fabricates an explanation?
- Grounded tools (especially Perplexity with citations) are much less likely to agree with a false premise, because the real sources will contradict it.

**Intermediate Note:** This "false premise" test is actually a standard evaluation technique used inside AI companies. When you're evaluating whether to ship an AI feature, this is one of the first tests you run.
