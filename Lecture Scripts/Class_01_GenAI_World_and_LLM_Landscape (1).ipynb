## Part A: Tokens — The Atoms of AI Language

> "LLMs don't read words. They read **tokens**. A token is a chunk of text — sometimes a word, sometimes part of a word, sometimes just a character.
>
> **The analogy: LEGO bricks.**
>
> When you write 'I love programming,' you see 3 words. The LLM sees something like: `[I]` `[ love]` `[ program]` `[ming]` — 4 tokens. It breaks text into LEGO-sized pieces it can work with.
>
> Why should you care? Because **tokens equal money and limits.** Every API call costs per token. Every model has a maximum number of tokens it can handle at once. If you're wasteful with tokens, you're burning cash and hitting walls.
>
> Think of it like SMS messages in the early 2000s — every character counted, and you learned to be concise. Same principle here, just at a larger scale."



![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/189/413/original/6.png?1776076487)


**Why tokens matter in practice:**

| Situation | Why Tokens Matter |
|-----------|-------------------|
| API billing | You pay per token (input + output) |
| Context limits | Model can only handle X tokens total |
| Response length | You can cap output tokens to control cost |
| Speed | More tokens = slower response |

**Beginner Note:** Don't stress about counting tokens manually. Just know: shorter, clearer prompts = fewer tokens = cheaper and faster.

````python
# Let's SEE tokens in action.
# We'll use the `tiktoken` library — the same tokenizer OpenAI uses.

!pip install tiktoken -q

import tiktoken

# This is the tokenizer used by GPT-4o
enc = tiktoken.encoding_for_model("gpt-4o")

# Try different inputs — see how token counts change
examples = [
    "Hello",
    "Hello, world!",
    "def fibonacci(n):\n    if n <= 1:\n        return n\n    return fibonacci(n-1) + fibonacci(n-2)",
    "Debugging Python code is frustrating",
    "SELECT u.name, o.total FROM users u JOIN orders o ON u.id = o.user_id WHERE o.total > 100",
]

print("=" * 70)
print(f"{'Text (preview)':<40} {'Words':<8} {'Tokens':<8} {'Ratio'}")
print("=" * 70)

for text in examples:
    tokens = enc.encode(text)
    words = len(text.split())
    preview = text[:37].replace('\n', ' ') + ("..." if len(text) > 37 else "")
    ratio = f"{len(tokens)/max(words,1):.1f}x"
    print(f"{preview:<40} {words:<8} {len(tokens):<8} {ratio}")

print("\nNotice: Code and SQL use significantly MORE tokens per word than plain English.")
print("This directly affects your API costs when building AI-powered features.")
````

````python
# Let's VISUALIZE what tokens actually look like.
# This shows you exactly how the model "sees" your text.

import tiktoken

enc = tiktoken.encoding_for_model("gpt-4o")

text = 'def calculate_total(items):\n    return sum(item.price for item in items)'

tokens = enc.encode(text)
decoded = [enc.decode([t]) for t in tokens]

print("Original code:")
print(text)
print(f"\nBroken into {len(tokens)} tokens:\n")

# Alternate markers for visibility
for i, token_text in enumerate(decoded):
    marker = "[A]" if i % 2 == 0 else "[B]"
    display = token_text.replace('\n', '<newline>').replace(' ', '<space>')
    print(f"  {marker} Token {i+1:>2}: '{display}'")

print("\nNotice how spaces, newlines, and underscores each become their own tokens.")
print("This is why formatting and whitespace in your prompts isn't free — it costs tokens.")
````

**Beginner Note:** See how the model doesn't see "words" — it sees chunks? Spaces, newlines, underscores all become separate pieces. This is one reason why the way you format your prompt matters.

## Part B: Context Window — The Model's Working Memory

> "Okay, so the model reads tokens. But how many tokens can it handle at once? That's the **context window**.
>
> **The analogy: Think of a desk.**
>
> Your context window is the desk size. Everything the model can 'see' — your prompt, the conversation history, AND its response — has to fit on this desk. Small desk? You can only work with a few pages. Huge desk? You can spread out an entire codebase.
>
> When you 'run out of context,' it's like your papers falling off the edge of the desk. The model literally forgets what was at the beginning of your conversation.
>
> Real-world example: ever had a long ChatGPT conversation where it starts contradicting what it said 30 messages ago? That's the desk getting full. Early messages fell off."



![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/189/416/original/7.png?1776077273)



**Context windows in 2026:**

| Model | Context Window | Real-World Equivalent |
|-------|---------------|----------------------|
| GPT-4o | 128K tokens | ~300 pages / one novel |
| Claude 3.5 Sonnet | 200K tokens | ~500 pages / two novels |
| Gemini 1.5 Pro | 1M tokens | ~2,500 pages / ten novels |
| Llama 3.1 | 128K tokens | ~300 pages |
| Mistral Large | 128K tokens | ~300 pages |




**Real-world impact:** Some engineering teams use Gemini's 1M-token context to feed entire microservice codebases into a single prompt and ask architectural questions. With GPT-4o's 128K, the same team has to be selective about which files they include. Neither is wrong — it's about knowing your desk size and working within it.

---

**Intermediate Note:** This is why long chat sessions "degrade." As the conversation grows, old messages eat into the context, leaving less room for the model to reason about your current question.



````python
# Context Window Calculator — How much fits on your "desk"?

import tiktoken

enc = tiktoken.encoding_for_model("gpt-4o")

# Simulate real-world content sizes
content_examples = {
    "A short prompt (bug fix request)": "Fix the bug in my login function that causes a 500 error when the email field is empty.",
    "A README file (~200 lines)": "# Project Name\n" * 200,
    "A Python file (~200 lines)": "def function_name(param):\n    return param\n" * 200,
    "A full chat session (~50 turns)": "User: Can you help me with this?\nAssistant: Of course! Here's what I suggest...\n" * 50,
}

context_windows = {
    "GPT-4o (128K)": 128_000,
    "Claude 3.5 (200K)": 200_000,
    "Gemini 1.5 (1M)": 1_000_000,
}

print("How much of each model's context window does this content use?\n")
print(f"{'Content':<38} {'Tokens':<10} {'GPT-4o':<14} {'Claude':<14} {'Gemini'}")
print("=" * 90)

for name, content in content_examples.items():
    token_count = len(enc.encode(content))
    gpt_pct = f"{token_count/128_000*100:.2f}%"
    claude_pct = f"{token_count/200_000*100:.2f}%"
    gemini_pct = f"{token_count/1_000_000*100:.3f}%"
    print(f"{name:<38} {token_count:<10} {gpt_pct:<14} {claude_pct:<14} {gemini_pct}")

print("\nTakeaway: Individual files barely dent the context window.")
print("But a long chat session with back-and-forth adds up fast.")
print("This is why starting a fresh conversation often gives better results.")
````

## Part C: Temperature — The Creativity Dial

> "Last concept — and this one's fun.
>
> **Temperature** controls how creative or predictable the model is. It's a number, usually between 0 and 1.
>
> **The analogy: A chef making a dish.**
>
> Temperature 0 = A chef following a recipe EXACTLY. Measure every gram. Same dish every time. Great for code generation — you want THE correct answer, not a creative interpretation of correctness.
>
> Temperature 1 = A jazz chef. Improvising. 'Let's throw in some sriracha and see what happens.' More creative, more varied, but occasionally... inedible.
>
> For software engineering, you almost always want low temperature. You want deterministic, reliable code — not a model getting 'creative' with your database queries.
>
> Real-world example: GitHub Copilot runs at very low temperature by default. When autocompleting code, you want predictability. But tools like Jasper.ai for marketing copy run at higher temperatures because variety is the point."

![image](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/189/426/original/8.png?1776078178)

**When to use which temperature:**

| Temperature | Use Case | Why |
|-------------|----------|-----|
| **0.0** | Code generation, bug fixes, SQL queries | You want THE correct answer, deterministically |
| **0.1 - 0.3** | Technical docs, test cases, refactoring | Mostly precise, tiny variation for natural language |
| **0.5 - 0.7** | Brainstorming, design discussions, naming | Balance of focus and creativity |
| **0.8 - 1.0** | Creative writing, exploring ideas, generating alternatives | Maximum variety |

**Beginner Note:** Most chat interfaces (ChatGPT, Claude.ai) don't let you set temperature directly — they pick a sensible default. You'll control temperature when calling APIs, which we'll do in later classes.

````python
import random
import math

def simulate_temperature(prompt_context, options_with_probs, temperature, num_samples=10):
    print(f"\nTemperature: {temperature}")
    print(f'Context: "{prompt_context} ___"\n')

    for i in range(num_samples):
        if temperature == 0:
            chosen = max(options_with_probs, key=lambda x: x[1])[0]
        else:
            scaled = [
                (w, math.exp(math.log(max(p, 1e-9)) / temperature))
                for w, p in options_with_probs
            ]

            total = sum(s for _, s in scaled)
            normalized = [(w, s / total) for w, s in scaled]

            r = random.random()
            cumulative = 0

            for w, p in normalized:
                cumulative += p
                if r <= cumulative:
                    chosen = w
                    break

        print(f'Run {i+1:02d}: "{prompt_context} {chosen}"')


# Flatter distribution (better for temperature demo)
code_options = [
    ("return user", 0.25),
    ("print(user)", 0.18),
    ("raise NotFound()", 0.15),
    ("yield user", 0.12),
    ("pass", 0.10),
    ("...", 0.08),
    ("return None", 0.07),
    ("log.debug(user)", 0.05),
]

print("=" * 60)
print("Scenario: What comes after 'def get_user(id):'?")
print("=" * 60)

random.seed(42)

for temp in [0, 0.3, 0.7, 1.0]:
    simulate_temperature("def get_user(id):", code_options, temp)
````

**The takeaway:** At temperature 0, it's "return user" every single time — deterministic and safe. At temperature 1, the model rolls the dice and sometimes picks unlikely options. For code, you want that consistency. For brainstorming, you want the variety.

# SECTION 6 — MINI BUILD: CHOOSING THE RIGHT MODEL FOR THE JOB

**Time: 25 minutes** | **Goal:** Students understand that different model tiers have dramatically different token costs and capabilities — and learn to match the right model to the right task.

> "In Section 5, we compared outputs across ChatGPT, Claude, and Gemini. Now let's go one level deeper — because here's something most beginners miss:
>
> **Not all models cost the same.** Behind every AI platform, there are multiple model tiers — frontier models that are powerful but expensive, and smaller models that are faster and cheaper but less capable.
>
> Think of it like cloud compute. You wouldn't run a simple health-check endpoint on a 64-core GPU server. And you wouldn't run your ML pipeline on a micro server. Same principle applies to LLMs: **right-size the model to the task.**
>
> You can't always switch models manually on free tiers — but when you start building with APIs (which we'll do later in this course), this becomes the single biggest lever for controlling costs. Let me show you why."

## The Model Tiers You Need to Know

Every major AI provider offers models at different capability levels. Here's how the landscape looks as of 2026:

| Tier | Examples | Quality | Speed | Cost |
|------|---------|---------|-------|------|
| **Frontier** | GPT-4o, Claude 3.5 Sonnet, Gemini 1.5 Pro | Highest — deep reasoning, subtle analysis | Moderate | Highest |
| **Mid-range** | GPT-4o mini, Claude Haiku, Gemini Flash | Good — handles most everyday tasks well | Fast | 10-20x cheaper |
| **Lightweight** | GPT-3.5 Turbo, older/smaller models | Basic — simple tasks, formatting, lookups | Fastest | 5-10x cheaper |



When you use ChatGPT's free tier, the platform decides which model to route your request to. When you use the API (which we'll cover in later classes), **you choose the model explicitly** — and you pay per token based on that choice.

**Real-world parallel:** This is the same tiering you see everywhere in infrastructure — S3 Standard vs S3 Glacier for storage, dedicated vs shared CPU for compute, premium vs economy for shipping. The capability is different, the price is different, and the right choice depends on the job.

## Why the Same Tokens Cost Different Amounts

> "Remember tokens from Section 4? Every API call costs money per token — both input and output. But here's the key insight most people miss: **different models charge wildly different rates for processing the same tokens.**
>
> A 500-token prompt sent to GPT-4o costs roughly 15 times more than the same 500 tokens sent to GPT-4o mini. The tokens are identical. The processing power behind them is not.
>
> This is the same economics as cloud compute. A t2.micro and a p4d.24xlarge both run Linux, both accept the same SSH connection — but one costs $0.01/hour and the other costs $32/hour. You pick based on the workload."

### Token Pricing Comparison (approximate, 2026)

| Model | Input Cost (per 1M tokens) | Output Cost (per 1M tokens) | Relative Cost |
|-------|---------------------------|----------------------------|---------------|
| GPT-4o | ~$2.50 | ~$10.00 | Baseline (1x) |
| GPT-4o mini | ~$0.15 | ~$0.60 | ~15x cheaper |
| GPT-3.5 Turbo | ~$0.50 | ~$1.50 | ~7x cheaper |
| Claude 3.5 Sonnet | ~$3.00 | ~$15.00 | ~1.3x more |
| Gemini 1.5 Pro | ~$1.25 | ~$5.00 | ~2x cheaper |

*Prices change frequently. The ratios matter more than the exact numbers.*

**Notice:** output tokens cost 2-4x more than input tokens on every model. This means a verbose AI response costs significantly more than a concise one — which is why prompt design directly affects your bill.

````python
# Let's make this concrete.
# How much would different tasks cost across different model tiers?

import tiktoken

enc = tiktoken.encoding_for_model("gpt-4o")

# Three common developer tasks at different complexity levels
tasks = {
    "Quick question (HTTP 429 meaning)": {
        "input": "What does HTTP status code 429 mean and when would I see it?",
        "est_output_tokens": 80
    },
    "Code review (50-line function)": {
        "input": "Review this function for bugs and improvements: " + "def process(data):\n    result = []\n    for item in data:\n        result.append(item)\n" * 12,
        "est_output_tokens": 500
    },
    "Architecture question (system design)": {
        "input": "Design a notification system for a mobile app handling 500K users with push, in-app, and email. Include queue, DB, and scaling strategy for 3 engineers.",
        "est_output_tokens": 1200
    }
}

# Pricing per 1M tokens (approximate, 2026)
models = {
    "GPT-4o":       {"input": 2.50, "output": 10.00},
    "GPT-4o mini":  {"input": 0.15, "output": 0.60},
    "GPT-3.5":      {"input": 0.50, "output": 1.50},
}

print("COST PER SINGLE REQUEST — same task, different model tiers")
print()
print(f"{'Task':<42} {'GPT-4o':<14} {'4o mini':<14} {'GPT-3.5':<14} {'Savings'}")
print("=" * 100)

for task_name, task_data in tasks.items():
    input_tokens = len(enc.encode(task_data["input"]))
    output_tokens = task_data["est_output_tokens"]

    costs = {}
    for model_name, pricing in models.items():
        cost = (input_tokens * pricing["input"] / 1_000_000) + \
               (output_tokens * pricing["output"] / 1_000_000)
        costs[model_name] = cost

    gpt4o_cost = costs["GPT-4o"]
    mini_cost = costs["GPT-4o mini"]
    savings = f"{gpt4o_cost/mini_cost:.0f}x cheaper with mini"

    print(f"{task_name:<42} ${gpt4o_cost:<12.5f} ${mini_cost:<12.5f} ${costs['GPT-3.5']:<12.5f} {savings}")

print()
print("=" * 100)
print()
print("Now scale it: 100 requests/day for a month (3,000 requests)...")
print()

for task_name, task_data in tasks.items():
    input_tokens = len(enc.encode(task_data["input"]))
    output_tokens = task_data["est_output_tokens"]
    monthly = 3000

    gpt4o_mo = monthly * ((input_tokens * 2.50 / 1_000_000) + (output_tokens * 10.00 / 1_000_000))
    mini_mo = monthly * ((input_tokens * 0.15 / 1_000_000) + (output_tokens * 0.60 / 1_000_000))

    print(f"{task_name[:38]:<40} GPT-4o: ${gpt4o_mo:>7.2f}/mo    4o mini: ${mini_mo:>6.2f}/mo")

print()
print("The per-request cost looks trivial. The monthly bill tells a different story.")
print("This is why model selection matters in production.")
````


The per-request difference is fractions of a cent. But at 100 requests/day, the monthly gap between GPT-4o and GPT-4o mini is the difference between 55/month and $3/month — and for many of those tasks, the smaller model is perfectly adequate.

**Real-world parallel:** This is exactly how engineering teams manage cloud costs. You don't run every service on the biggest instance. You right-size: small instances for lightweight services, big instances for heavy compute. Same discipline for LLMs.

**The 80/20 rule in practice:** In most developer workflows, roughly 80% of AI interactions are medium-complexity tasks — "explain this error," "write a test for this function," "summarize this PR." A mid-range model handles these well. Save frontier models for the 20% that actually needs deep reasoning.

## Token Optimization — Getting Better Output for Fewer Tokens

Beyond choosing the right model, **how you write the prompt** directly affects how many tokens you use — and therefore your cost, speed, and rate limits.

| Habit | Why It Helps | Example |
|-------|-------------|----------|
| **Be specific in your prompt** | Vague prompts produce long, rambling outputs | "List 3 risks" vs "Tell me about risks" (the second generates 3x more output tokens) |
| **Set a length constraint** | Prevents the model from over-explaining | "Answer in under 100 words" or "Maximum 5 bullet points" |
| **Ask for structured output** | Tables and lists use fewer tokens than paragraphs | A 5-row table uses ~60% of the tokens of the same info written as prose |
| **Don't repeat context unnecessarily** | Every token in the input costs money | In long conversations, summarize previous context instead of re-pasting everything |
| **Request only what you need** | Output tokens cost 2-4x more than input tokens | "Return only the function" vs "Explain your approach and then write the function" |

**Beginner Note:** On the free tiers of ChatGPT, Claude, and AI Studio, you aren't charged per token — you hit rate limits instead. But these habits still matter because they make responses faster, more focused, and easier to use. And when you start building with APIs, they directly affect your bill.

**Advanced Note:** In production systems, teams often use a "model router" — a lightweight classifier that reads the incoming request and routes simple tasks to cheap models and complex tasks to expensive ones. Same pattern as a CDN routing static assets to edge servers and dynamic requests to origin.

````python
# How much do length constraints affect token usage?

import tiktoken

enc = tiktoken.encoding_for_model("gpt-4o")

# Two versions of the same question -- one vague, one constrained
prompts = {
    "Vague (no constraints)": "Explain what a REST API is and how it works.",
    "Constrained (specific)": "Explain what a REST API is in exactly 3 bullet points, each under 15 words.",
}

# Estimated output tokens based on typical model behavior
estimated_outputs = {
    "Vague (no constraints)": 350,
    "Constrained (specific)": 80,
}

print("PROMPT DESIGN vs TOKEN USAGE")
print()
print(f"{'Prompt Style':<30} {'Input Tokens':<15} {'Est. Output':<15} {'Total'}")
print("=" * 70)

results = {}
for name, prompt in prompts.items():
    input_t = len(enc.encode(prompt))
    output_t = estimated_outputs[name]
    total = input_t + output_t
    results[name] = total
    print(f"{name:<30} {input_t:<15} {output_t:<15} {total}")

vague = results["Vague (no constraints)"]
constrained = results["Constrained (specific)"]
reduction = (1 - constrained / vague) * 100

print(f"\nThe constrained prompt uses ~{reduction:.0f}% fewer total tokens.")
print("Same information. Better format. Lower cost. Faster response.")
print("\nThis is why prompt engineering isn't just about quality -- it's about efficiency.")
````

> "Let me tie this together. In the last 25 minutes, you learned three things that most people don't figure out until months into using AI:
>
> First, **not all models are priced equally.** Frontier models like GPT-4o cost 15-20x more per token than mid-range models like GPT-4o mini. Same tokens in, very different bill.
>
> Second, **the right model depends on the task complexity.** Simple tasks don't need a frontier model. Complex reasoning tasks justify the cost. This is cloud right-sizing applied to AI.
>
> Third, **how you write the prompt directly affects token usage.** A specific, constrained prompt produces shorter, more focused output — which means fewer tokens, lower cost, and faster responses.
>
> On the free tiers you're using today, this is about speed and rate limits. When you start building with APIs, it's about money. Either way, these habits are worth building now."
